## Build Linear Regression Model

### Define Features and Target

```python
selected_features = ['open_thea', 'budget', 'runtime', 'MPAArating',
                     'Director_gross','Actor_gross','studio_gross',
                     'Sci-Fi', 'Adventure', 'Action', 'Animation', 'Musical', 'Comedy',
                     'Fantasy', 'Thriller', 'Horror', 'Drama', 'Romance', 'Historical',
                     'Western', 'Family', 'Crime', 'Concert', 'Documentary', 'Sports', 'War',
                     'mean_score_mtc','comment_count_mtc',
                     'trend_a_week', 'trend_two_week', 'trend_a_day',
                     'hot_season', 'tier-a-series', 'strong_competition']
target = 'open_gross'
```

```python
X = df[selected_features].values
y=list(df['open_gross'])
```

```python
X.shape
```

```python
y
```

### Let's Train-test split the data!

```python
# import train test split function from sklearn
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Linear Regression model using sklearn

```python
# Import linear regression from sklearn
from sklearn.linear_model import LinearRegression

# Initialize model by providing parameters
# http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
reg = LinearRegression()

# Fit a model by providing X and y from training set
reg.fit(X_train, y_train)

# Make prediction on the training data
y_train_pred = reg.predict(X_train)

# Make predictions on test data
y_test_pred = reg.predict(X_test)
```

### Calculate the metric scores for the model

```python
# Import metrics functions from sklearn
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, y_test_pred)
```

```python
# setup the figure and axes
fig = plt.figure(figsize=(5, 5))
movie_id=[y.index(i) for i in y_test]
# movie_id
plt.scatter(movie_id,y_test,label="Test")
# plt.show()
plt.scatter(movie_id,y_test_pred,label="Predicted")
plt.legend()
plt.show()
```

## Build Decision Tree

### Decision Tree model using sklearn

```python
# Import Decision Tree from sklearn
from sklearn.tree import DecisionTreeRegressor

# Fit regression model
reg = DecisionTreeRegressor(max_depth=5)

# Fit a model by providing X and y from training set
reg.fit(X_train, y_train)

# Make prediction on the training data
y_train_pred = reg.predict(X_train)

# Make predictions on test data
y_test_pred = reg.predict(X_test)
```

### Calculate the metric scores for the model

```python
mean_absolute_error(y_test, y_test_pred)
```

```python
# setup the figure and axes
fig = plt.figure(figsize=(5, 5))
movie_id=[y.index(i) for i in y_test]
# movie_id
plt.scatter(movie_id,y_test,label="Test")
# plt.show()
plt.scatter(movie_id,y_test_pred,label="Predicted")
plt.legend()
plt.show()
```
